##Criação do Modelo de Regressão Linear Múltipla Manual

```python
class RegressaoLinearMultipla:
    '''
    Uma classe que implementa o modelo de regressão linear com gradiente descendente.
    '''
    def __init__(self, taxa_aprendizado=0.01, iteracoes=100):
        self.taxa_aprendizado = taxa_aprendizado
        self.iteracoes = iteracoes
        self.pesos, self.bias = None, None
        self.erros = []
        
    @staticmethod 
    def _erro_quadratico_medio(y, y_hat):
        '''
        Método privado, usado para avaliar a perda em cada iteração.
        
        :parâmetro: y - array, valores verdadeiros
        :parâmetro: y_hat - array, valores previstos
        :return: float
        '''
        erro = 0
        for i in range(len(y)):
            erro += (y[i] - y_hat[i]) ** 2
        return erro / len(y)
    
    def fit(self, X, y):
        '''
        Usado para calcular o coeficiente do modelo de regressão linear.
        
        :parâmetro X: array (preditoras)
        :parâmetro y: array (classes)
        :retorna: None
        '''
        #Inicializar pesos e bias com zeros
        self.pesos = np.zeros(X.shape[1])
        self.bias = 0
        
        #Realizar a descida do gradiente
        for i in range(self.iteracoes):
            #Equação da reta
            y_hat = np.dot(X, self.pesos) + self.bias
            erros = self._erro_quadratico_medio(y, y_hat)
            self.erros.append(erros)
            
            #Calcula as derivadas
            pesos_parciais = (1 / X.shape[0]) * (2 * np.dot(X.T, (y_hat - y)))
            derivadas_parciais = (1 / X.shape[0]) * (2 * np.sum(y_hat - y))
            
            #Atualiza os coeficientes 
            self.pesos -= self.taxa_aprendizado * pesos_parciais
            self.bias -= self.taxa_aprendizado * derivadas_parciais
        
        
    def predict(self, X):
        '''
        Faz previsões usando a equação da linha.
        
        :parâmetro X: array (recursos)
        :retorna: array, (previsões)
        '''
        return np.dot(X, self.pesos) + self.bias
```

```python
model=RegressaoLinearMultipla()
```

```python
model.fit(X,y)
```

```python
xs = np.arange(len(model.erros))
ys = model.erros

plt.plot(xs, ys, lw=3, c='#087E8B')
plt.title('Erro por iteração (MSE)', size=20)
plt.xlabel('Iteração', size=14)
plt.ylabel('Erro', size=14)
plt.show()
```

```python
erros = {}
for taxa_aprendizado in [0.5, 0.1, 0.01, 0.001]:
    model = RegressaoLinearMultipla(taxa_aprendizado=taxa_aprendizado)
    model.fit(X, y)
    erros[f'taxa_aprendizado={str(taxa_aprendizado)}'] = model.erros
```

```python
xs = np.arange(len(model.erros))

plt.plot(xs, erros['taxa_aprendizado=0.5'], lw=3, label=f"taxa_aprendizado = 0.5, Final = {erros['taxa_aprendizado=0.5'][-1]:.2f}")
plt.plot(xs, erros['taxa_aprendizado=0.1'], lw=3, label=f"taxa_aprendizado = 0.1, Final = {erros['taxa_aprendizado=0.1'][-1]:.2f}")
plt.plot(xs, erros['taxa_aprendizado=0.01'], lw=3, label=f"taxa_aprendizado = 0.01, Final = {erros['taxa_aprendizado=0.01'][-1]:.2f}")
plt.plot(xs, erros['taxa_aprendizado=0.001'], lw=3, label=f"taxa_aprendizado = 0.001, Final = {erros['taxa_aprendizado=0.001'][-1]:.2f}")
plt.title('Perda por iteração (MSE) para diferentes taxas de aprendizagem', size=20)
plt.xlabel('Iteração', size=14)
plt.ylabel('Erro', size=14)
plt.legend()
plt.show()
```

###Sum of Squares Due to Regression (SSR)

```python
SSR =np.sum(np.square(y - y_previstos))
SSR
```

###Residual Standard Error (RSE)

```python
 rse = np.sqrt(SSR / (len(y_previstos) - 2))
 rse
```

###Sum of Squares Total (SST)

```python
SST = sum((y-np.mean(y))**2)  
SST
```

###R2

```python
R2=1 - (float(SSR))/SST
R2
```

###R2 Ajustado

Cálculo Manual

```python
R2_Ajustado=1 - (1-R2)*(len(y)-1)/(len(y)-X.shape[1]-1)
```

```python
print(f'R2: {R2.round(4)}\nR2 Ajustado: {R2_Ajustado.round(4)}')
```

Cálculo com Sklearn

```python
print(f'R2: {model.score(X, y).round(4)}\nR2 Ajustado: {(1 - (1-model.score(X, y))*(len(y)-1)/(len(y)-X.shape[1]-1)).round(4)}')
```

###Erros Padronizados

```python
N = len(df_X)
p = len(df_X.columns) + 1  # Adicionamos um devido o termo do interceptor

X_with_intercept = np.empty(shape=(N, p), dtype=np.float)
X_with_intercept[:, 0] = 1
X_with_intercept[:, 1:p] = df_X.values

beta_hat = np.linalg.inv(X_with_intercept.T @ X_with_intercept) @ X_with_intercept.T @ df_y.values
print(beta_hat)
```

```python
y_hat = model.predict(X)
residuals = df_y.values - y_hat
residual_sum_of_squares = residuals.T @ residuals
sigma_squared_hat = residual_sum_of_squares / (N - p)
var_beta_hat = np.linalg.inv(X_with_intercept.T @ X_with_intercept) * sigma_squared_hat
for p_ in range(p):
    standard_error = var_beta_hat[p_, p_] ** 0.5
    print(f"SE(beta_hat[{p_}]): {standard_error}")
```
